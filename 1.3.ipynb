```python
import pandas as pd
import numpy as np

# Creating a DataFrame with one numeric and one text column
df = pd.DataFrame({
    "ID": np.arange(1, 6),
    "Value": np.random.randint(10, 100, size=5),
    "Name": ["Divya", "Rajeev", "Coco", "Merika", "Raymika"]
})

df
```

```python
#Using the lambda function
df["Value_Double"] = df["Value"].apply(lambda x: x * 2)
df
```

```python
#Creating a flag column
df["StartsWithA"] = df["Name"].apply(lambda x: x.startswith("D"))
df
```
